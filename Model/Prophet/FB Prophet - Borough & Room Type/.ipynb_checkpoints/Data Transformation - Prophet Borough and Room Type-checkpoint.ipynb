```python
import pandas as pd
```

```python
calendar_2017 = pd.read_csv('calendar_2017.csv',error_bad_lines=False)
calendar_2018 = pd.read_csv('calendar_2018.csv',error_bad_lines=False)
calendar_2019 = pd.read_csv('calendar_2019.csv',error_bad_lines=False)
calendar_2020 = pd.read_csv('calendar_2020.csv',error_bad_lines=False)
calendar_2021a = pd.read_csv('calendar_2021a.csv',error_bad_lines=False)
calendar_2021b= pd.read_csv('calendar_2021b.csv',error_bad_lines=False)
```

```python
calendar_frames = [calendar_2017, calendar_2018, calendar_2019, calendar_2020, calendar_2021a, calendar_2021b]
calendar_combined = pd.concat(calendar_frames)
calendar_combined
```

```python
calendar_combined['price'] =calendar_combined['price'].str.replace('$', '')
calendar_combined['price'] = calendar_combined['price'].str.replace(',', '')
```

```python
calendar_combined['price'] = calendar_combined['price'].astype(float)
```

```python
calendar_combined['date']= pd.to_datetime(calendar_combined['date'])
```

```python
calendar_combined = calendar_combined[['date', 'listing_id','price']]
calendar_combined
```

```python
listings_master = pd.read_csv("listings-masters.csv", low_memory=False)
listings_master
```

```python
calendar_combined = calendar_combined.rename(columns={'listing_id': 'id'})
```

```python
calendar_final = pd.merge(calendar_combined, listings_master, on="id", how="inner")
calendar_final
```

```python
calendar_final = calendar_final[['date', 'borough_id', 'room_typeid','price_x']]
calendar_final
```

```python
calendar_final = calendar_final.rename(columns={"price_x": "price"})
```

```python
calendar_final = pd.DataFrame(calendar_final.groupby(["date","borough_id", "room_typeid"])["price"].mean().reset_index())
calendar_final
```

```python
calendar_final.to_csv('calendar_borough.csv')
```
