```python
import pandas as pd
```

### **2017**

```python
df = pd.read_csv('2017.csv')
df
```

```python
calendar_2017 = df.loc[df['date'] < "2018-01-01"]
```

```python
calendar_2017 = calendar_2017.loc[calendar_2017['available'] == "t"]
```

```python
calendar_2017=calendar_2017.dropna()
```

```python
calendar_2017
```

```python
calendar_2017.to_csv("calendar_2017.csv")
```
