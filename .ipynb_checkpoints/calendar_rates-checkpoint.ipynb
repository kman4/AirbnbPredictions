```python
import pandas as pd
import datetime as dt
```

```python
calendar = pd.read_csv("calendar.csv", error_bad_lines=False)
calendar
```

```python
calendar.dtypes
```

```python
calendar['date']= pd.to_datetime(calendar['date'])
```

```python
calendar_2021 = calendar.loc[calendar['date'] < "2022-1-1"]
calendar_2021
```

```python
calendar_2021['month'] = calendar_2021['date'].dt.month_name()
calendar_2021
```

```python
calendar_2021['day_of_week'] = calendar_2021['date'].dt.day_name()
calendar_2021
```

```python
calendar_2021['price'] = calendar_2021['price'].str.replace('$', '')
calendar_2021['price'] = calendar_2021['price'].str.replace(',', '')
calendar_2021
```

```python
calendar_2021['price'] = calendar_2021['price'].astype(float)
```

```python
listings_master = pd.read_csv("listings-master.csv", low_memory=False)
listings_master
```

```python
calendar_2021 = calendar_2021.rename(columns={'listing_id': 'id'})
```

```python
calendar_2021_merge = pd.merge(calendar_2021, listings_master, on="id", how="inner")
```

```python
calendar_2021_merge
```

```python
rates_monthly = pd.DataFrame(calendar_2021_merge.groupby([ "borough_id", "month"])["price_x"].mean().reset_index())
rates_monthly
```

```python
rates_monthly.dtypes
```

```python
rates_dayOfWeek = pd.DataFrame(calendar_2021_merge.groupby([ "borough_id", "day_of_week"])["price_x"].mean().reset_index())
rates_dayOfWeek
```

```python
rates_monthly.to_csv("rates_monthly.csv")
```

```python
rates_dayOfWeek.to_csv("rates_dayOfWeek.csv")
```
